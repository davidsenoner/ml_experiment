```python
from sklearn.linear_model import SGDRegressor
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 2*np.random.rand(100, 1)
y = 4+3*X+np.random.randn(100, 1)
```

```python
regressor = SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, eta0=0.1)
regressor.fit(X, y.ravel())  # training the model

score = regressor.score(X, y.ravel())

print("R-squared:", score)
print(regressor.intercept_, regressor.coef_)
```

```python
X_test = np.array([[0], [2]])
X_test_b = np.c_[np.ones((2,1)), X_test]

prediction = X_test_b.dot(np.array([regressor.intercept_, regressor.coef_]))  # prediction using numpy
prediction_1 = regressor.predict(X_test) # predition using SGDRegresor method

print("Prediction Numpy:", prediction)
print("Prediction SGDRegression:", prediction_1)
```

```python
# plot
plt.title("SGDRegessor Prediction")
plt.scatter(X, y, marker='.')
plt.plot(X_test, prediction_1, "r-")
# plt.plot(X_test, prediction, "b-")
plt.ylabel("Prediction")
plt.xlabel("Test values")
```
