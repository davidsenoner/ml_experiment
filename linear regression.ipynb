Linear Regression using numpy array operations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 2*np.random.rand(100, 1)
y = 4+3*X+np.random.randn(100, 1)

plt.scatter(X, y, marker='.')
```

```python
X_b = np.c_[np.ones((100,1)), X]
mat = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print(mat) # minimized cost function using numpy
```

```python
# inference
X_test = np.array([[0], [2]])
X_test_b = np.c_[np.ones((2,1)), X_test]

prediction = X_test_b.dot(mat)

print(prediction)
```

```python
# plot
plt.scatter(X, y, marker='.')
plt.plot(X_test, prediction, "r-")
plt.ylabel("Prediction")
plt.xlabel("Test values")
```

Linear Regerssion using sklearn

```python
from sklearn.linear_model import LinearRegression

lin_regression = LinearRegression()
lin_regression.fit(X, y)  # fit using initial training values
lin_regression.intercept_, lin_regression.coef_
```

```python
# predict using fitted model

lin_regression.predict(X_test)
```

Test with scipy methods

```python
mat2, residuals, rank, s = np.linalg.lstsq(X_b, y, rcond=1e-6)
```

Moore–Penrose inverse A +

```python
mat3 = np.linalg.pinv(X_b).dot(y)
```

```python
print(f"Numpy: {mat}")
print(f"Scipy: {mat2}")
print(f"sklearn: {lin_regression.intercept_, lin_regression.coef_}")
print(f"Moore–Penrose inverse (pinv): {mat3}")
```
